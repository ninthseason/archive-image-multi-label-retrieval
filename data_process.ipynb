```python
import pandas as pd
```

```python
train_data = pd.read_csv("data/train_label.csv", header=None)
train_data.head()
```

```python
train_data.shape
```

```python
train_label_string = train_data[1]
train_label_list = train_label_string.map(lambda x: eval(x))
train_label_length = train_label_list.map(lambda x: len(x))
print(f"标签数量分布:\n{train_label_length.value_counts()}")
```

```python
print(2)

train_class_set = set()
for i in train_label_list:
    for j in i:
        train_class_set.add(j)
print(f"训练集类别数: {len(train_class_set)}")
```

```python
train_class_list = list(train_class_set)
category_map = {label: i for i, label in enumerate(train_class_list)}
category_map
```

```python
import json
with open("category_map.json", "w", encoding="utf-8") as f:
    json.dump(category_map, f)
```

```python
with open("category_map.json", "r") as f:
    print(json.load(f))
```

```python
train_data
```

```python
import numpy as np
from tqdm import tqdm

def getLabelVector(categories):
    label = np.zeros(4375)
    # label_num = len(categories)
    for c in categories:
        print(c)
        index = category_map[str(c)]
        label[index] = 1.0  # / label_num
    return label


l = train_data
labels = []
for i in tqdm(range(l.shape[0])):
    categories = eval(l[1][i])
    label = getLabelVector(categories)
    labels.append(label)
    break
```

```python
labels_np = np.array(labels)
np.save("train_label_vectors.npy", labels_np)
```

```python
test_data = pd.read_csv("data/test_query.csv", header=None)
test_label_list = test_data[0].map(lambda x: eval(x))

test_class_set = set()
for i in test_label_list:
    for j in i:
        test_class_set.add(j)

print(f"测试集类别数: {len(test_class_set)}")
```

```python
all_class_set = train_class_set.union(test_class_set)
print(f"全集类别数: {len(all_class_set)}")
print(f"测试集-训练集: {len(test_class_set.difference(train_class_set))}")
```
