```python
import numpy as np
import pandas as pd
test_query = pd.read_csv("data/test_query.csv", header=None)
test_query = test_query[0]
test_query = test_query.map(lambda x: eval(x))

train_label = pd.read_csv("data/train_label.csv", header=None)
train_label = train_label[1]
train_label = train_label.map(lambda x: eval(x))

train_label_set = set()
for i in train_label:
    for j in i:
        train_label_set.add(j)
train_label_list = list(train_label_set)
```

```python
import synonyms
```

```python
r = synonyms.compare("穿着", "衣服", seg=False)
r
```

```python
from tqdm import tqdm

def nearby(word):
    nearby_w = ""
    maxvalue = 0
    for w in tqdm(train_label_list):
        value = synonyms.compare(word, w, seg=False)
        if value > maxvalue:
            maxvalue = value
            nearby_w = w
    if maxvalue > 0.8:
        return nearby_w
    else:
        return ""

nearby("衣着")
```

```python
for iidx, i in enumerate(test_query):
    for idx, j in enumerate(i):
        if j not in train_label_set:
            i[idx] = nearby(j)
    test_query[iidx] = i
```

```python
test_query_set = set()
for i in test_query:
    for j in i:
        test_query_set.add(j)
```

```python
print(test_query_set.difference(train_label_set))
```
